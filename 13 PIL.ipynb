```python
# open image with PIL
from PIL import Image
im=Image.open('./img/sample01a.jpg') # 'open' means take an image and store to a variable
print(im.format, im.size, im.mode)
```

```python
# im.save('filename.ext')
im.save('sample01a-png1.png')
im2=Image.open('./sample01a-png1.png')
im2.format
```

```python
# im.save('filename','ext')
im.save('sample01a-png2','jpeg')
im3=Image.open('./sample01a-png2')
im3.format
#Note: explorer will show that its type File
```

```python
im.save('sample01a-png3.png','jpeg')
im4=Image.open('./sample01a-png3.png')
im4.format
#Note: explorer will show that its type PNG
```

```python
#圖片quality 調整：
#JPEG可設定Quality= 1 - 100
#PNG 可設定一維數值作為透明度
im5 = Image.open('./img/sample01a.jpg')
im5.save('sample01a-100jpg.jpg',quality = 100)
```

```python
im6 = Image.open('./img/img_0499.jpg')
a = im6.getdata() # 此為抓取圖檔的原始資料
list1=list(a)
list1
```

```python
# 取出圖像Meta Data
exif = im6._getexif() #a dictionary of metadata, but using number code as keys
from PIL.ExifTags import TAGS
for key,value in exif.items():
    print(TAGS.get(key),":",value) #presenting as tags
```

```python
from glob import glob
from os.path import splitext
jpglist=glob('./img/dataset/covid/*.[jJ][pP][eE][gG]')  
if jpglist==[]:
    print('目錄或檔案的副檔名錯誤')
else:
    for jpg1 in jpglist:
        im5=Image.open(jpg1)
        print('-'*70)
        #print(im5)
        print(splitext(jpg1)[0]) # 擷取出圖像檔案的主檔名
        png1=splitext(jpg1)[0]+'.png' #圖檔的主檔名加上.png 
        im5.save(png1) #儲存為png格式
        print(png1)
```
