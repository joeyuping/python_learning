# Linear Regression

for predicting quantitative result == essentially finding the formula that best match x to y

```python
import numpy as np
import pandas as pd
import sklearn.linear_model as slm

temperatures = np.array([[29,28,34,31,25,29,32,31]]).transpose()   # note 2D array : [sample1[feature1,feature2],sample2[feature1,feature2]]
drink_sales = np.array([7.7,6.2,9.3,8.4,5.9,6.4,8.0,7.1])

print('建立線性迴歸物件')
lm = slm.LinearRegression().fit(temperatures,drink_sales) # fit straight line
print('迴歸係數(gradient)：', lm.coef_)
print('截距（y-intercept)：', lm.intercept_)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt

plt.scatter(temperatures,drink_sales)
drink_sales_fit = lm.coef_ * temperatures + lm.intercept_
plt.plot(temperatures, drink_sales_fit,'r')
plt.xlabel('temperatures')
plt.ylabel('drink_sales')
plt.show()
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt

plt.scatter(temperatures,drink_sales)
drink_sales_fit = lm.predict(temperatures)
new_temperature = np.array([[34,40,20,35]]).transpose()
drink_sales_predict = lm.predict(new_temperature)
plt.scatter(new_temperature, drink_sales_predict, c='g',marker='x')
plt.plot(temperatures, drink_sales_fit,'r')
plt.xlabel('temperatures')
plt.ylabel('drink_sales')
plt.show()
```
