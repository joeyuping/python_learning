pro - Skimage works directly from numpy array to image
* skimage is solely python, openCV supports python, c++, java.
* if skimage has the same function as openCV, skimage tends to be faster as skimage is written directly with python
* because skimage works with numpyarray directly, most numpy array function works with skimage

## read image

```python
from PIL import Image
from skimage import io # use io to open image files
img1 = Image.open('./img/amanda.jpg')
img2 = io.imread('./img/amanda.jpg')
print(img2) # see that img2 is indeed a numpy array
print(img1.__class__, img2.__class__) 
print(img1.size, img2.size) # PIL- image dimension, Skimage - size of numpy array - number of elements
print(img2.shape) # Skimage - shape of the numpy array
```

## show image

```python
import matplotlib.pyplot as plt

from skimage import data
img = data.chelsea()

io.imshow(img)  # basically same as pyplot!!!
plt.axis('off') # see that io is in fact a pyplot object!
io.show()
```

## other methods/attributes

```python
print(img.shape)
print(img[0][0]) # pixel info a H-0, W-0
print(img[0,0,:]) # alternative syntax
io.imsave('cat1.jpg',img)
```

## extract RGB channel

```python
print('R:\n',img[:,:,0])
print('G:\n',img[:,:,1])
print('B:\n',img[:,:,2])
```

```python
R = img.copy()
R[:,:,1:] = 0

G = img.copy()
G[:,:,0] = 0
G[:,:,2] = 0

B = img.copy()
B[:,:,:2] = 0

fig,ax = plt.subplots(1,3, constrained_layout=True)
ax[0].imshow(R)
ax[1].imshow(G)
ax[2].imshow(B)
io.show()
```
