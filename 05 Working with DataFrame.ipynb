### (generating data set)

```python
import pandas as pd
import numpy as np

data = {
    '縣市': ['台北','台北','台北','新竹','新竹','新竹'], 
    '年份': [2000,2001,2002,2001,2002,2003],
    '人口': [2.2,2.5,2.7,1.1,1.5,2.0]
}

df = pd.DataFrame(data, 
                     columns=['縣市','人口','年份'])

df_customindex = pd.DataFrame(data, 
                     columns=['縣市','人口','年份'], 
                     index=['one','two','three','four','five','six'])
display(df.head())
display(df_customindex.head())
```

## Selecting rows, columns, cell

### selecting columns

```python
# use []
display(df['縣市'])

# use [list(labels)]
display(df[['縣市','年份']]) # note list of columns

# use iloc
display(df.iloc[:,1:2])
#note: can not use  df[:3] to select the first three columns! this default to selecting rows!

# use loc
display(df.loc[:,['縣市','年份']])
```

### selecting rows

```python
#selecting the first 4 rows

display(df.head(4)) # if left blank is first 5 rows

display(df[:4])  # note: 4 NOT included!

display(df.iloc[:4])  # note: 4 NOT included!

display(df.loc[:4]) # note: 4 included! the 4 here is the index "label"

#select particular rows
display(df.iloc[[0,4]])  #note: list of indexes

display(df.tail(4))
```

```python
# if custom index:

df_customindex.loc[['one','three']]

df_customindex.iloc[2:6]
```

### selecting cell
* df.loc[row label, column label]
* df.iloc[row indexes, column indexes]

```python
display(df)

#using .loc[row index slice,column index slice]
display(df.loc[:3,['縣市','年份']])  #note: "list" of column names / indexes within the frame/loc index
```

```python
#using iloc
display(df.iloc[:4 , 1:2])

display(df.iloc[2:4 , :]) 
```

```python
#using reindex
display(df_customindex.reindex(index=['one','four'],
              columns = ['縣市','年份']))
```

## extracting rows, columns from dataframe into series

```python
# extract vertically or horizontally from DataFrame, both will result in a serie!

display(df['縣市']) # extracting vertically one column

print()

display(df_customindex.loc['one']) # extracting horizontally one row, (column names become the index)
```

```python
#note! when assigning a slice of the frame to a variable, its only a reference! not true copying!
Series1 = df_customindex['縣市']
Series1['one'] = '花蓮'
print(Series1)

print()

print(df_customindex)
# see that Series1 is only a reference to the column. altering the content of the column will alter to the original frame!
```

### to do true copying, do this:

```python
Serial1 = df_customindex['縣市'].copy()
Serial1['one'] = '屏東'
print(Serial1)
print()
print(df_customindex)
# see that a change to Serial1 no longer alters the original frame!
```

```python
# same goes for ndarray
arr1 = np.array([1,2,3])
arr2 = arr1[1:3]
print("arr2:",arr2)
arr2[1] = 4
print("arr2:",arr2)
print("arr1:",arr1)
#see that assigning a slice of ndarray to a variable also only takes a reference!
```

```python
# true copying for ndarray
arr1 = np.array([1,2,3])
arr2 = arr1[1:3].copy()  # note here!! .copy()
print("arr2:",arr2)
arr2[1] = 4
print("arr2:",arr2)
print("arr1:",arr1)
```

## Alter column value

```python
data = {
    '縣市': ['台北','台北','台北','新竹','新竹','新竹'], 
    '年份': [2000,2001,2002,2001,2002,2003],
    '人口': [2.2,2.5,2.7,1.1,1.5,2.0]
}

frame = pd.DataFrame(data, 
                     columns=['縣市','人口','年份','負債'],  # there is no '負債' in data!
                     index=['one','two','three','four','five','six'])
frame  # see how column '負債' is added with NAN as all its values while creating a DataFrame
```

```python
# set the value for whole column
frame['負債'] = 1.6
frame
```

```python
frame['負債'] = np.arange(6.)
frame
```

```python
frame['負債'] = np.random.randint(6,50,6) #(low,high,size)
frame
```

```python
frame['負債'] = np.random.rand(6) * 100 #(size), generate float
frame
```

```python
ndArray = np.random.rand(6) * 100
frame['負債'] = ndArray.astype(np.int64) #convert value to integer
frame
```

## Add new column

```python
# how to add new column? just define a new column!
empty = np.empty(6)
empty.fill(np.nan)
frame['新開發區'] = empty
display(frame)
#to create new column and fill will empty values seems not so easy!


frame['新開發區'] = [True,True,False,True,False,False] # directly fill the new column with value
display(frame) # use display(frame) to present the df
```

## Delete Column/Row - drop(), del

```python
# using drop with label and axis
# drop columns
display(frame.drop('新開發區', axis = 1))  

# drop row
display(frame.drop('two', axis = 0))

display(frame) #Note: drop has "inplace" default to faulse - will not mutate original
```

```python
# drop using index= , columns=
frame.drop(columns = '新開發區',index=['four','six'])
```

```python
# drop inplace = true
display(frame.drop(columns = '新開發區',inplace = True)) 
# when inplace is True, return nothing, as the original frame has been altered

display(frame)
```

```python
# using del - only for column
del frame['負債'] # this mutate the original directly
```

## sum(), mean()

```python
# generate data

import numpy as np
import pandas as pd

scores = np.random.randint(50,101,(50,5))
scores_df = pd.DataFrame(scores, columns = ['國文','英文','數學','地理','歷史'],
                        index = range(1,51))
scores_df.index.name = '座號'
scores_df.columns.name = '科目'
scores_df.head()
```

```python
# doing sum(), mean()

print(scores_df.sum())
print(scores_df.mean())

#note: default is sum according to columns (axis: 0)
#output is a serie
```

```python
#usage
for index in range(1,51):
    print("學號{:s} : 總分 {:d} , 平均{:2.1f}".format(str(index).zfill(2),
                                                scores_df.loc[index].sum(),
                                                scores_df.loc[index].mean()))
# note usage of   str(index).zfill(2)  to fill leading zeros
# scores_df.loc[index] to extract row
```

```python
Series_sum = scores_df.sum(axis=1) # axis 1 is the horizontal
Series_mean = scores_df.mean(axis=1) # output is a serie
print(Series_sum)
print(Series_mean)
```

```python
# adding the above into a new column:
scores_df['總分'] = Series_sum
scores_df['平均'] = Series_mean
scores_df.head()
```
