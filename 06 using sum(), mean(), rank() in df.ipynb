# using sum(), mean(), rank() in df

```python
import numpy as np
import pandas as pd

scores = np.random.randint(50,101,[10,5])
scores_pd = pd.DataFrame(scores, index = range(1,11), columns=['國文','英文','數學','地理','自然'])
scores_pd
```

```python
sum_values = scores_pd.sum(axis=1)
mean_values = scores_pd.mean(axis=1)
rank_values = sum_values.rank(method='min')
rank_values
```

```python
scores_pd['總分'] = sum_values
scores_pd['平均'] = mean_values
scores_pd['排名'] = rank_values
```

```python
scores_pd
```

```python
# doing rank inline

if '排名' in scores_pd.columns: 
    del scores_pd['排名']

scores_pd['排名'] = scores_pd['總分'].rank(method='min',ascending=False)
scores_pd
```

```python
scores_pd.sort_values('排名')
```
