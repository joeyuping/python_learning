```python
from matplotlib import pyplot as plt
img = plt.imread('./img/amanda.jpg')
#print(img) # notice img is also a numpy array
print(img[0,0,:])  # can work with ndarray just like in scikit 
print(img.size)
print(img.shape)
plt.imshow(img)
plt.show()

# almost exact same functionality as skimage! but lack of image manipulation functions
```

## saving image
plt.savefig('file.jpg') ==> save the plt figure with axis and everything, default 680,420  
plt.imsave('file.png', image) ==> save the full size image

```python
plt.savefig('fig.jpg') # note, has to happen before plt.show()
plt.imsave('image.png', img)
```
