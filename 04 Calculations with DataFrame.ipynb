# Calculations with DataFrame

* df.add(df2, fill_value=0)   fill_value = the value to fill in size different and need to fill

### (generating data)

```python
import pandas as pd
import numpy as np
df1 = pd.DataFrame(np.arange(12.).reshape(3,4), columns=list('abcd'))
df2 = pd.DataFrame(np.arange(20.).reshape(4,5), columns=list('abcde'))
display(df1)
display(df2)
```

## Addition

```python
df1 + df2

# note: additions only occur at intersections of the two df. 
# size follows the larger one
# as empty, so fill with NaN
# NaN plus anything is still NaN
```

```python
# df.add(df2) equivalent to df1 + df2, but with support to do fill_value

df1.add(df2, fill_value=0)
```

## same for sub, mul, div

```python
df1.sub(df2, fill_value=0)
```

```python
df1.mul(df2, fill_value=2)
```

```python
df1.div(df2, fill_value=5)
```
