# NOTE !!!  openCV read  BGR, not RGB!!! color will flip!

```python
import cv2
from skimage import io

img = cv2.imread('./img/rgb.jpg')
img2 = io.imread('./img/rgb.jpg')

import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,2, constrained_layout = True)

ax[0].imshow(img) 
ax[0].set_title('cv2 - BGR')
ax[1].imshow(img2)
ax[1].set_title('skimage - RGB')
plt.show()
```

## converting BGR to RGB

```python
img = cv2.cvtColor(cv2.imread('./img/amanda.jpg'), cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```
