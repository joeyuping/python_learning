## using sklearn.linear_model.LinearRegression().fit
### with sklearn.preprocessing import PolynomialFeatures
### to do Polynomial Regression
pro - has predict

```python
import numpy as np
import sklearn.linear_model as slm

y = [8,6,5,8,10,14,17]
x = np.arange(len(y)).reshape(-1,1)
x_plot = np.linspace(0,6,100).reshape(-1,1)

from sklearn.preprocessing import PolynomialFeatures
poly1=PolynomialFeatures(degree=2)   # to change degree of poly, just change here! 
# notice that when degree is high, it leads to over fit and serious anomaly

poly1x = poly1.fit_transform(x)

print(poly1x) # note output of poly1x = matrix of 1,x,x^2...

lm = slm.LinearRegression().fit(poly1x,y)

print(lm.coef_)  # a0,a1,a2
print(lm.intercept_)
y_fit = np.dot(poly1x,lm.coef_)+lm.intercept_  # method 1 - using poly1x matrix, the formula become same as straight line 


polyx_plot = poly1.fit_transform(x_plot)
y_fit2 = lm.predict(polyx_plot)  # method 2

import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x_plot,y_fit2)
plt.show()
```

## analysing the effectiveness of model using 殘值
plot y_original - y_predicted

```python
import seaborn as sns

y_error = y - y_fit
sns.displot(y_error, kind='kde') # kde = kernel density estimation
```

## accessing the quality of model using various kinds of error measurement

```python
from sklearn.metrics import r2_score, mean_squared_error, median_absolute_error, mean_absolute_error, mean_squared_log_error
import numpy as np

print('MSE:',sum(y_error**2)/len(y))
print('MSE:',mean_squared_error(y,y_fit))

print('RMSE:',np.sqrt(sum(y_error**2)/len(y)))
print('RMSE:',np.sqrt(mean_squared_error(y,y_fit)))


# if mean absolute error differs alot from median absolute error, ther might be significant outlier
print('mean_absolute_error:',sum(np.abs(y-y_fit))/len(y))
print('mean_absolute_error:',mean_absolute_error(y,y_fit))

print('median_absolute_error:',np.median(np.abs(y-y_fit)))
print('median_absolute_error:',median_absolute_error(y,y_fit))

# coefficient of determination, how well does the model describe the data as compared to simply taking mean of y
# usually about >0.75 is acceptable
print('mean_squared_log_error:', mean_squared_log_error(y,y_fit))
print('r2_score:', r2_score(y,y_fit))  


# Mean square error
# Root mean square
```

## multi feature regression

```python
# say ,how waist and height could determine weight

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

waist_heights = np.array([[67,160],[68,165],[70,167],[65,170],[80,165],[85,167],[78,178],[79,182],[95,175],[89,172]])
weights = np.array([50,60,65,65,70,75,80,85,90,81])
lm = LinearRegression().fit(waist_heights, weights)
weights_fit = lm.predict(waist_heights)

#setup figure
%matplotlib notebook
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel('waist')
ax.set_ylabel('height')
ax.set_zlabel('weight')


# plot data
ax.scatter(waist_heights[:,0],waist_heights[:,1],weights.reshape(-1,1))

# plot best fit plane
x,y = np.meshgrid(np.linspace(65,95,2),np.linspace(160,190,2))
z = lm.coef_[0]*x + lm.coef_[1]*y + lm.intercept_
z2 = lm.coef_*np.array([x,y]) + lm.intercept_
ax.plot_surface(x,y,z,color='r',alpha=0.1)

plt.show()
```

## looking at linearity of each feature 

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

waist_heights = np.array([[67,160],[68,165],[70,167],[65,170],[80,165],[85,167],[78,178],[79,182],[95,175],[89,172]])
weights = np.array([50,60,65,65,70,75,80,85,90,81])

fig,axs = plt.subplots(1,3,constrained_layout = True)
axs[0].scatter(waist_heights[:,0],waist_heights[:,1])
axs[0].set_xlabel('waist')
axs[0].set_ylabel('height')
t0= 'corr:' + str(np.round(np.corrcoef(waist_heights[:,0],waist_heights[:,1])[0][1],2))
axs[0].set_title(t0)

axs[1].scatter(waist_heights[:,0],weights)
axs[1].set_xlabel('waist')
axs[1].set_ylabel('weights')
t1= 'corr:' + str(np.round(np.corrcoef(waist_heights[:,0],weights)[0][1],2))
axs[1].set_title(t1)

axs[2].scatter(waist_heights[:,1],weights)
axs[2].set_xlabel('height')
axs[2].set_ylabel('weights')
t2= 'corr:' + str(np.round(np.corrcoef(waist_heights[:,1],weights)[0][1],2))
axs[2].set_title(t2)

fig.show()
```

## using np.polyfit
con - to plot graph have to type out the whole formula

```python
import numpy as np
import matplotlib.pyplot as plt

y = [8,6,5,8,10,14,17]
x = np.arange(len(y))

plt.scatter(x,y)
plt.title("test")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
a3, a2, a1, a0 = np.polyfit(x,y,3)
y2 = [a3 * i ** 3 + a2 * i ** 2 + a1 * i + a0  for i in x]
plt.plot(x,y2,"--")
plt.plot(x,y,"rx")
plt.show()
```

```python
#polyfit ploter

x = [1,2,3,4,5,6,7]
y = [2,6,5,8,10,14,17]

def polyfitY(x,y,fitDegree):
    coeff = np.array(np.polyfit(x,y,fitDegree))
    y2 = np.array([])
    for i in x:
        polyx = np.array([])
        for j in range(fitDegree, -1 ,-1):
            polyx = np.append(polyx, i ** j)
        y2 = np.append(y2, np.dot(polyx,coeff))
    return y2

plt.plot(x, y, "rx")
plt.plot(x, polyfitY(x,y,1))
plt.show()
    
```
